```python
# User-defined function for covariance

def get_covariance(dataframe):
  n = dataframe.shape[0]
  col_set = set(dataframe.columns)
  index_col = 'Columns'
  cov_matrix = pd.DataFrame(columns=[index_col]+list(col_set), index=[i for i in range(0,len(col_set))])
  i = 0
  for x in list(col_set):
    cov_matrix.loc[i,index_col] = x
    x_mean = dataframe[x].mean()
    for y in list(col_set):
      y_mean = dataframe[y].mean()
      x_values = dataframe[x].values.tolist()
      y_values = dataframe[y].values.tolist()
      prod = 0
      for xi, yi in zip(x_values, y_values):
        prod += (xi-x_mean)*(yi-y_mean)
      cov_x_y = prod/n
      cov_x_y = "{:.6e}".format(cov_x_y)
      cov_matrix.loc[i,y] = cov_x_y
    i += 1
  cov_matrix.set_index(index_col,inplace=True)
  return cov_matrix 

covar_dataf = get_covariance(df)
covar_dataf
```

```python
# correlation user-defined function

def get_correlation(dataframe):
  n = dataframe.shape[0]
  col_set = set(dataframe.columns)
  index_col = 'Columns'
  corr_matrix = pd.DataFrame(columns=[index_col]+list(col_set), index=[i for i in range(0,len(col_set))])
  i = 0
  for x in list(col_set):
    corr_matrix.loc[i,index_col] = x
    x_mean = dataframe[x].mean()
    x_stdv = dataframe[x].std()
    for y in list(col_set):
      y_mean = dataframe[y].mean()
      y_stdv = dataframe[y].std()
      x_values = dataframe[x].values.tolist()
      y_values = dataframe[y].values.tolist()
      prod = 0
      for xi, yi in zip(x_values, y_values):
        prod += (xi-x_mean)*(yi-y_mean)
      cov_x_y = prod/n
      # Pearson's correlation coefficient = covariance(X, Y) / (stdv(X) * stdv(Y))
      corr_x_y = cov_x_y/(x_stdv*y_stdv)

      corr_matrix.loc[i,y] = corr_x_y

    i += 1

  corr_matrix.set_index(index_col,inplace=True)

  return corr_matrix 
```
